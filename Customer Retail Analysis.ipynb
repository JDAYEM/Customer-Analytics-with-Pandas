# Average time between purchases for each customer

### Order by data set by customers and dates 

```python
df_avg =  df.sort_values(by = ['Customer_Name' , 'Date'] ,  ascending = False)
```

```python
df_avg.tail(10)
```

### get previous purchase dates for each customer -----------Lagging 

```python
df_avg ['Previous Purchase Date'] = df_avg.groupby('Customer_Name')['Date'].shift(-1)
```

```python
df_avg.head(5)
```

### Cast dates columns 

```python
df_avg[['Date' , 'Previous Purchase Date']] =  df_avg[['Date' , 'Previous Purchase Date']].apply(pd.to_datetime)
```

### Time between purchases 

```python
df_avg['DiFF BT Purchases'] = df_avg['Date'] -  df_avg['Previous Purchase Date']
```

```python
df_avg.head(50)
```

### Convert Nat Values to datetime

```python
df_avg['DiFF BT Purchases'] = pd.to_timedelta(df_avg['DiFF BT Purchases'] )
```

### Fill Nat with 0

```python
df_avg['DiFF BT Purchases'].fillna(pd.Timedelta('0 days') , inplace = True)
```

###  Convert Date time to integers 

```python
df_avg['DiFF BT Purchases'] = df_avg['DiFF BT Purchases'].dt.days
```

```python
df_avg.head(50)
```

### Average time between purchases for each customer 

```python
Avg_time_customer = df_avg.groupby('Customer_Name')['DiFF BT Purchases'].mean().reset_index()
```

```python
Avg_time_customer
```

### Average time between purchases across all transactions

```python
Avg_time_all = df_avg['DiFF BT Purchases'].mean()
```

# Customer Purchases compare to Average

#### custom function to get average status

```python
def avg_status ( amount  ,  avg):
    if amount > avg:
        return 'Above Average'
    elif amount < avg:
        return 'Below Average'
    else:
        return 'Exactly'
```

```python
df['Avg'] = df.total_paid.mean()
df.Avg
```

```python
df['pur_status'] = 'M'
```

```python
df.dtypes
```

```python
df['Avg'] = df.Avg.astype(int)
```

```python
df.head(5)
```

```python
df['Diff BT Avg Tot'] =  df.total_paid -  df.Avg
```

#### for loop to calc pur status 

```python
for i in df.index:
    df['pur_status'].iloc[i] = avg_status ( df['total_paid'].iloc[i]  ,  df['Avg'].iloc[i]  ) 
```
